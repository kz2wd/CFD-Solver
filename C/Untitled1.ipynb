```python
import numpy as np
import pyqtgraph as pg
from pyqtgraph.Qt import QtCore

N = 256
img = pg.image(np.zeros((N, N)))

x = np.linspace(0, 4*np.pi, N)
X, Y = np.meshgrid(x, x)
t = 0

def update():
    global t
    field = np.sin(X + t) * np.cos(Y + t)
    img.setImage(field, autoLevels=False)
    t += 0.1

timer = QtCore.QTimer()
timer.timeout.connect(update)
timer.start(16)
```
